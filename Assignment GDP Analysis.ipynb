```python
# Imputing the null values with mean values
z.loc[np.isnan(z['Himachal Pradesh']), ['Himachal Pradesh']] = z['Himachal Pradesh'].mean()
z.loc[np.isnan(z['Maharashtra']), ['Maharashtra']] = z['Maharashtra'].mean()
z.loc[np.isnan(z['Manipur']), ['Manipur']] = z['Manipur'].mean()
z.loc[np.isnan(z['Mizoram']), ['Mizoram']] = z['Mizoram'].mean()
z.loc[np.isnan(z['Nagaland']), ['Nagaland']] = z['Nagaland'].mean()
z.loc[np.isnan(z['Punjab']), ['Punjab']] = z['Punjab'].mean()
z.loc[np.isnan(z['Rajasthan']), ['Rajasthan']] = z['Rajasthan'].mean()
z.loc[np.isnan(z['Tripura']), ['Tripura']] = z['Tripura'].mean()

# Plotting the graph 
plt.figure(figsize=(15, 5))
z.mean().sort_values().plot.bar()
plt.xlabel('State')
plt.ylabel('Average % Growth over previous year')
plt.show()

```

```python
x=df.drop('Item', axis=1)
primary = (((x.iloc[[6]].sum())/(x.iloc[[30]]))*100).rename(index={30:'Primary'})
secondary=(((x.iloc[[10]].sum())/(x.iloc[[30]]))*100).rename(index={30:'Secondary'})
tertiary=(((x.iloc[[26]].sum())/(x.iloc[[30]]))*100).rename(index={30:'Tertiary'})

prim_secon= primary.append(secondary)
sectors= prim_secon.append(tertiary)
```

```python
# Plotting The graph
per_gdp = sectors.transpose()
plt.figure(figsize=(15,5))
per_gdp.plot.bar(stacked=True)
plt.xlabel('States')
plt.ylabel('Sectors as a percentage of GDP')
plt.title('percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP for all the states')
plt.show()
```

```python
# Quantile Values to categorise data in C1, C2, C3, C4
converted_arr = [float(x) for x in list(df.iloc[32, 1:])]
quant_20 = np.quantile(converted_arr, 0.2)
quant_50 = np.quantile(converted_arr, 0.5)
quant_85 = np.quantile(converted_arr, 0.85)
quant_1 = np.quantile(converted_arr, 1.0)

# Getting the quantile values
print("C1", quant_1)
print("C2", quant_85)
print("C3", quant_50)
print("C4", quant_20)
```

```python
# For ease of analysis transposing the dataframe
df_trans= df.transpose()
df_trans.head()
```

```python
df_trans.columns =  list(df_trans.iloc[0,:])
df_trans.drop('Item', inplace=True)
df_trans.head()
```

```python
# Changing the Datatypes from object to numeric types.
final_df =  pd.DataFrame(index= df_trans.index)
for column in (df_trans.columns):
    final_df[column] = [float(x) for x in df_trans[column]]
final_df.head()
```

```python
# Dividing Data Set into a categories according to quantiles
C1 = final_df.loc[final_df['Per Capita GSDP (Rs.)'] >= quant_85]
C2 = final_df.loc[(final_df['Per Capita GSDP (Rs.)'] >= quant_50) & (final_df['Per Capita GSDP (Rs.)']< quant_85)]
C3 = final_df.loc[(final_df['Per Capita GSDP (Rs.)'] >= quant_20) & (final_df['Per Capita GSDP (Rs.)']< quant_50)]
C4 = final_df.loc[(final_df['Per Capita GSDP (Rs.)'] < quant_20)] 
```

```python
# Removing unwanted columns from the set
c1 = C1.drop(['Primary', 'Secondary', 'Tertiary', 'TOTAL GSVA at basic prices', 'Taxes on Products','Subsidies on products','Gross State Domestic Product', "Population ('00)",'Per Capita GSDP (Rs.)'], axis=1)
c2 = C2.drop(['Primary', 'Secondary', 'Tertiary', 'TOTAL GSVA at basic prices', 'Taxes on Products','Subsidies on products','Gross State Domestic Product', "Population ('00)",'Per Capita GSDP (Rs.)'], axis=1)
c3 = C2.drop(['Primary', 'Secondary', 'Tertiary', 'TOTAL GSVA at basic prices', 'Taxes on Products','Subsidies on products','Gross State Domestic Product', "Population ('00)",'Per Capita GSDP (Rs.)'], axis=1)
c4 = C2.drop(['Primary', 'Secondary', 'Tertiary', 'TOTAL GSVA at basic prices', 'Taxes on Products','Subsidies on products','Gross State Domestic Product', "Population ('00)",'Per Capita GSDP (Rs.)'], axis=1)
```

```python
# Plotting the graph for C1
plt.figure(figsize=(15,10))
s1 = pd.Series(c1.sum()/C1['Gross State Domestic Product'].sum())*100
s1.sort_values().plot.barh()
plt.ylabel('Sub Sectors')
plt.xlabel('Percentage of Total GSDP')
plt.title('C1')
plt.show()
```

```python
# Plotting the Graph for C2
plt.figure(figsize=(15,10))
s2 = pd.Series(c2.sum()/C2['Gross State Domestic Product'].sum())*100
s2.sort_values().plot.barh()
plt.ylabel('Sub Sectors')
plt.xlabel('Percentage of Total GSDP')
plt.title('C2')
plt.show()
```

```python
# Plotting Graph for C3
plt.figure(figsize=(15,10))
s2 = pd.Series(c3.sum()/C3['Gross State Domestic Product'].sum())*100
s2.sort_values().plot.barh()
plt.ylabel('Sub Sectors')
plt.xlabel('Percentage of Total GSDP')
plt.title('C3')
plt.show()
```

```python
# Plotting Graph for C3
plt.figure(figsize=(15,10))
s2 = pd.Series(c4.sum()/C4['Gross State Domestic Product'].sum())*100
s2.sort_values().plot.barh()
plt.ylabel('Sub Sectors')
plt.xlabel('Percentage of Total GSDP')
plt.title('C4')
plt.show()
```

```python
# Corelation of sub sectors with GSDP
relation =  final_df.drop(['Primary', 'Secondary', 'Tertiary', 'TOTAL GSVA at basic prices', 'Taxes on Products','Subsidies on products', "Population ('00)",'Per Capita GSDP (Rs.)'], axis=1)
cor = relation.corr()
plt.figure(figsize=(15,10))
sns.heatmap(cor, cmap="YlGnBu", annot=True)
plt.show()
```
